```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_csv("/content/drive/MyDrive/Predicting yield of maize in Visakhapatnam.csv")
df1
```

```
df2=df1.drop(['Unnamed: 0','Crop'],axis='columns')
df2
```

```
import seaborn as sns
import matplotlib.pyplot as plt
cor = df2.corr()
plt.figure(figsize = (12,10))

top_corr_features=cor.index
g=sns.heatmap(df2[top_corr_features].corr(), annot=True, cmap ="RdYlGn")
```

```
df3=df2.drop(['Area','SEP','AUG','JUN','mansoon'],axis='columns')
df3
```

```
# Features that are most important in determining yield of Maize in Visakhapatnam
O=df3.drop(['Production'],axis='columns')
l=df3[['Production']]
# from sklearn.model_selection import train_test_split
# O_train,O_test,l_train,l_test=train_test_split(O,l,test_size=0.3)
```

```
from sklearn.linear_model import LinearRegression
ma=LinearRegression()
ma.fit(O,l)
mal=ma.predict(O)
LR=ma.score(O,l)
print(ma.predict(O))
print(LR)
df4=pd.DataFrame(mal,columns=['linearmaize'])
df4
```

```
from sklearn.svm import SVR
mas=SVR(C=80,gamma="auto",kernel='linear')
mas.fit(O,l)
SV=mas.score(O,l)
premas=mas.predict(O)
print(premas)
print(SV)
df5=pd.DataFrame(premas,columns=['svrmaize'])
df5
```

```
from sklearn.neighbors import KNeighborsRegressor 
mak= KNeighborsRegressor(n_neighbors=2, metric='minkowski', p=5 )  
mak.fit(O,l)
knmak= mak.predict(O)
KNN=mak.score(O,l)
print(knmak)
print(mak.score(O,l))
df6=pd.DataFrame(knmak,columns=['knnmaize'])
df6
```

```
from sklearn.ensemble import RandomForestRegressor
maize=RandomForestRegressor(n_estimators=10)
maize.fit(O,l)
RAN=maize.score(O,l)
premaize=maize.predict(O)
print(maize.score(O,l))
df7=pd.DataFrame(premaize,columns=['ranmaize'])
df7
```

```
import matplotlib.pyplot as plt
m=df4['linearmaize']
n=df2['Production']
o=df5['svrmaize']
p=df6['knnmaize']
q=df7['ranmaize']
x=df2['Crop_Year']
plt.plot(x,m,label="Linear Regression",color='yellow',marker="")
plt.plot(x,n,label="Actual",color='red',marker="")
plt.plot(x,o,label="SVR",color='green',marker="")
plt.plot(x,p,label="KNN",color='black',marker="")
plt.plot(x,q,label="RandomForest",color='blue',marker="")
# plt.xlim(0,20)
# plt.ylim(0,30000,5000)
# plt.axis([0,20,0,30000])
plt.xlabel("year")
plt.ylabel("production")
plt.title("(a) Performance of various algorithms of maize",fontsize=20)
plt.legend()
plt.savefig("Performance of various algorithms of maize.jpg")
plt.show()
```
