```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_csv("/content/drive/MyDrive/Predicting yield of Rice in Visakhapatnam.csv")
df1
```

```
df2=df1.drop(['Unnamed: 0','Crop'],axis='columns')
df2
```

```
import seaborn as sns
import matplotlib.pyplot as plt
cor = df2.corr()
plt.figure(figsize = (12,10))

top_corr_features=cor.index
g=sns.heatmap(df2[top_corr_features].corr(), annot=True, cmap ="RdYlGn")
```

```
df3=df2.drop(['SEP'],axis='columns')
df3
```

```
# Features that are most important in determining yield of Rice in Visakhapatnam
E=df3.drop(['Production'],axis='columns')
f=df3[['Production']]
# from sklearn.model_selection import train_test_split
# E_train,E_test,f_train,f_test=train_test_split(E,f,test_size=0.3)
```

```
from sklearn.linear_model import LinearRegression
ril=LinearRegression()
ril.fit(E,f)

LR=ril.score(E,f)
print(LR)
preril=ril.predict(E)
print(ril.predict(E))

df4=pd.DataFrame(preril,columns=['linearrice'])
df4
```

```
from sklearn.svm import SVR
ris=SVR( C=1,gamma="scale",kernel='linear')
ris.fit(E,f)
SV=ris.score(E,f)
svrris=ris.predict(E)
print(svrris)
print(SV)
df5=pd.DataFrame(svrris,columns=['svrrice'])
df5
```

```
from sklearn.neighbors import KNeighborsRegressor 
rik= KNeighborsRegressor(n_neighbors=2, metric='minkowski', p=2 )  
rik.fit(E, f)
knrik= rik.predict(E)
KNN=rik.score(E,f)
print(knrik)
print(rik.score(E,f))
df6=pd.DataFrame(knrik,columns=['knnrice'])
df6
```

```
from sklearn.ensemble import RandomForestRegressor
Rice=RandomForestRegressor(n_estimators=40)
Rice.fit(E, f)
RAN=Rice.score(E,f)
prerice=Rice.predict(E)
print(RAN)
df7=pd.DataFrame(prerice,columns=['ranrice'])
df7
```

```
import matplotlib.pyplot as plt
m=df4['linearrice']
n=df2['Production']
o=df5['svrrice']
p=df6['knnrice']
q=df7['ranrice']
x=df2['Crop_Year']
plt.plot(x,m,label="Linear Regression",color='yellow',marker="")
plt.plot(x,n,label="Actual",color='red',marker="")
plt.plot(x,o,label="SVR",color='green',marker="")
plt.plot(x,p,label="KNN",color='black',marker="")
plt.plot(x,q,label="RandomForest",color='blue',marker="")
# plt.xlim(2000,2016,1000)
# plt.ylim(0,30000,5000)
# plt.axis([2000,2016,0,30000])
plt.xlabel("year")
plt.ylabel("production")
plt.title("(a) Performance of various algorithms of Rice",fontsize=20)
plt.legend()
plt.savefig("Performance of various algorithms of Rice.jpg")
plt.show()
```
