```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_csv("/content/drive/MyDrive/Predicting yield of Ragi in Visakhapatnam.csv")
df1
```

```
df2=df1.drop(['Unnamed: 0','Crop'],axis='columns')
df2
```

```
import seaborn as sns
import matplotlib.pyplot as plt
cor = df2.corr()
plt.figure(figsize = (12,10))

top_corr_features=cor.index
g=sns.heatmap(df2[top_corr_features].corr(), annot=True, cmap ="RdYlGn")
```

```
df3=df2.drop(['JUL','SEP','AUG'],axis='columns')
df3
```

```
# Features that are most important in determining yield of Ragi in Visakhapatnam
R=df3.drop(['Production'],axis='columns')
s=df3[['Production']]
# from sklearn.model_selection import train_test_split
# R_train,R_test,s_train,s_test=train_test_split(R,s,test_size=0.2)
```

```
from sklearn.linear_model import LinearRegression
# from sklearn.model_selection import cross_val_score
ral=LinearRegression()
# cross_val_score(ral,R,s,cv=3)
ral.fit(R,s)
LR=ral.score(R,s)
print(LR)
preral=ral.predict(R)

print(ral.predict(R))
# print(ral.score(R_test,s_test))
df4=pd.DataFrame(preral,columns=['linearragi'])
df4
```

```
from sklearn.svm import SVR
# from sklearn.model_selection import cross_val_score
ras=SVR(kernel='linear',C=10,gamma='scale')
# cross_val_score(ras,R,s,cv=3)
ras.fit(R,s)
SV=ras.score(R,s)
SV

print(ras.score(R,s))
svrras=ras.predict(R)
print(svrras)
df5=pd.DataFrame(svrras,columns=['svrragi'])
df5
```

```
from sklearn.neighbors import KNeighborsRegressor 
rak= KNeighborsRegressor(n_neighbors=2, metric='minkowski', p=2 )  
rak.fit(R, s)
knrak= rak.predict(R)
KNN=rak.score(R,s)
print(KNN)
print(knrak)
df6=pd.DataFrame(knrak,columns=['knnragi'])
df6
```

```
from sklearn.ensemble import RandomForestRegressor
Ragi=RandomForestRegressor(criterion='mse')
Ragi.fit(R, s)
RAN=Ragi.score(R,s)
print(RAN)
preragi=Ragi.predict(R)
df7=pd.DataFrame(preragi,columns=['ranragi'])
df7
```

```
import matplotlib.pyplot as plt
m=df4['linearragi']
n=df2['Production']
o=df5['svrragi']
p=df6['knnragi']
q=df7['ranragi']
x=df2['Crop_Year']
plt.plot(x,m,label="Linear Regression",color='yellow',marker="")
plt.plot(x,n,label="Actual",color='red',marker="")
plt.plot(x,o,label="SVR",color='green',marker="")
plt.plot(x,p,label="KNN",color='black',marker="")
plt.plot(x,q,label="RandomForest",color='blue',marker="")
# plt.xlim(0,20)
# plt.ylim(0,30000,5000)
# plt.axis([0,20,0,30000])
plt.xlabel("year")
plt.ylabel("production")
plt.title("(a) Performance of various algorithms of ragi",fontsize=20)
plt.legend()
plt.savefig("Performance of various algorithms of ragi.jpg")
plt.show()
```
